***Leitura do Arquivo CSV através do Pandas***

***Visualizar quais os valores de classificação que temos no dataset e suas respectivas quantidades.***

***Através do método pivot_table, podemos visualizar a classificação dos filmes em uma tabela de usuário x filme***

```python
matriz = ratings_small.pivot_table(index='userId', columns='movieId', values='rating')
matriz
```

***Trocar todos os valores Nan por 0 para que não haja problemas na hora de fazer a análise***

```python
matriz_nova = matriz.fillna(0)
matriz_nova
```

***Através de uma posição escolhida aleatoriamente da "matriz_nova", vamos atribuir um valor de 2 de forma que gere uma matriz B que é a matriz_nova com elementos "estragados"***

```python
import random
while True:
    i, j = random.randint(0, 670), random.randint(0, 9064)
    if matriz_nova.iloc[i, j] != 0:
        print(i, j)
        break
B = matriz_nova.values
print(B[i][j])
B[i][j] = 2
B
```

***Agora, vamos decompor a matriz B em componentes através da função svd da biblioteca scipy e obter as matrizes U, S e Vt. A matriz U é a matriz de autovetores da matriz B.B^T, a matriz S é a matriz diagonal de autovalores de B.B^T e a matriz Vt é a matriz de autovetores da matriz B^T.B. Com isso conseguimos fazer uma projeção PCA usando somente os auto-vetores ligados aos maiores auto-valores com intuito de melhor visualizar os dados com poucas distorções da matriz B.***

```python
from scipy.linalg import svd, diagsvd
import numpy as np
U, s, Vt = svd(B)
s = s
print(s[-1])
s[:100] *= 0


```

***Agora vamos recompor a matriz B com os auto-vetores e auto-valores que escolhemos anteriormente e comparar com a matriz_nova original. Com isso podemos visualizar a diferença entre o valor real da matriz_nova e o valor estimado pela matriz B.***

```python
sigma = diagsvd(s, U.shape[1], Vt.shape[0])
matriz_reconstruida = U.dot(sigma.dot(Vt))
matriz_reconstruida[i][j]
diferenca = abs(matriz_reconstruida[i][j] - B[i][j])
print(diferenca)
```

***A próxima etapa é fazer esse processo repetidas vezes escolhendo uma posição aleatória da matriz B e atribuindo um valor aleatório para ela, e com os valores resultantes do processo e do valor original da matriz, gerar um gráfico da diferença entre o valor real e o valor estimado pela matriz B para visualizar os erros cometidos. Para isso, vamos criar uma função que calcula a diferença entre o valor original e o valor estimado, a função recebe como parametros a matriz original e um valor K, que é quantos componentes vamos eliminar da nossa matriz de auto-valores.*** 

```python
def calcula_erro(B,k):
    while True:
        i, j = random.randint(0, 670), random.randint(0, 9064)
        if matriz_nova.iloc[i, j] != 0:
            novo_valor = random.randint(0, 5)
            break
    B[i][j] = novo_valor
    U, s, Vt = svd(B)
    s = s
    s[-k:] *= 0
    sigma = diagsvd(s, U.shape[1], Vt.shape[0])
    matriz_reconstruida = U.dot(sigma.dot(Vt))
    diferenca = abs(matriz_reconstruida[i][j] - B[i][j])
    return diferenca
```

```python
#criar um dataframe com colunas k e erro
df = pd.DataFrame(columns=['k', 'margem_erro'])
#salvar o dataframe em um arquivo csv
for i in range(2):
    for k in range(1, 671):
        erro = calcula_erro(B,k)
        df.loc[len(df)] = [k, erro]
        print(k, erro)

df.to_csv('margem_de_erros.csv', index=False)
```

```python
#criar um histograma com os valores de k e erro
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('margem_de_erros.csv')
df.plot(x='k', y='margem_erro', kind='hist', bins=100, title='Margem de erro para diferentes valores de k')
plt.show()
```
